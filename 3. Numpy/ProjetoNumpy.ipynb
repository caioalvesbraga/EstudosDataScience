### Fórmula para calcular coeficiente angular:

â = (n * sum(Xi*Yi) - sum(Xi) * sum(Yi))/(n * Soma(Xi^2) - (Soma(Xi))^2)

### Fórmula para calcular o coeficiente linear

b = mean(Yi) - a * mean(Xi)

# Analisando um Dataset (Oranges vs Grapefruit)

```python
import numpy as np
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
data = np.loadtxt(url, delimiter = ',', usecols = np.arange(1, 6, 1), skiprows = 1)

diametro_laranja = data[:5000, 0]
diametro_toranja = data[5000:, 0]
peso_laranja = data[:5000, 1]
peso_toranja = data[5000:, 1]

import matplotlib.pyplot as plt

plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
```

```python
## Calculando os coeficientes da Laranja
X = diametro_laranja
Y = peso_laranja
n = np.size(X)

a = (n * np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - sum(X)**2) ## Coeficiente Angular
b = np.mean(Y) - a * np.mean(X)

y = a*X+b
b
```

```python
# Calculando coeficiente angular com números aleatórios
b = 17
np.random.seed(84)
coef_angular = np.random.uniform(low=10, high=20, size=100)

norma = np.array([])
for i in range(100):
  norma = np.append(norma, np.linalg.norm(Y - coef_angular[i]*X + b))

np.min(norma) # Norma mais justa ou
coef_angular[norma == np.min(norma)] # Para o coefciente angular melhor ajustado pela norma
```
